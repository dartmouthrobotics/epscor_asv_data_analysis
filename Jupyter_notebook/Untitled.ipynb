```python
fig, ax = plt.subplots(1, 2, figsize=(12, 8))

for i, res in enumerate(['l', 'h']):
    m = Basemap(projection='gnom', lat_0=43.4095, lon_0=-72.038,
                width=500000, height=500000, resolution=res, ax=ax[i])
    m.fillcontinents(color="#FFDDCC", lake_color='#DDEEFF')
    m.drawmapboundary(fill_color="#DDEEFF")
    m.drawcoastlines()
    m.drawrivers()
    ax[i].set_title("resolution='{0}'".format(res));

    ax[i].set_xlabel('Longitude')
    ax[i].set_ylabel('Latitude')

    ax[i].plot(np.array([-72.033128]), np.array([43.4096079]), color='k', marker='o', markersize=13)
    ax[i].plot(np.array([-72.033128]), np.array([43.4096079]), color='yellow', marker='o', markersize=8)
    
    m.plot(np.array([-72.038]), np.array([43.4095]), color='k', marker='o', markersize=13)
    m.plot(np.array([-72.038]), np.array([43.4095]), color='yellow', marker='o', markersize=8)
    
```
