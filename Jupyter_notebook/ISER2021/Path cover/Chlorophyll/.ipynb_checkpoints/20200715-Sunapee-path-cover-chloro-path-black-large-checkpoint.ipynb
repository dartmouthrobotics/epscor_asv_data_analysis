# calculate distance

```python
from geopy.distance import distance
dist = 0 
for i in range(len(x)):
    if i != 0:
        coords_1_x = x[i]
        coords_1_y = y[i]
        coords_2_x = x[i-1]
        coords_2_y = y[i-1]
        coords_1 = (coords_1_x, coords_1_y)
        coords_2 = (coords_2_x, coords_2_y)
        dist += distance(coords_1, coords_2).m
    else:
        continue
dist
```
