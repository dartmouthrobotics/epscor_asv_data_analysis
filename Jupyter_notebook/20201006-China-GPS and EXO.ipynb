```python
# Load essential libraries
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.tri as tri
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from matplotlib import ticker, cm
import numpy as np
from numpy import ma
import csv

from datetime import datetime
```

```python
#!/usr/bin/env python
degree_sign= u'\N{DEGREE SIGN}'

CSV_FILE_PATH = '../Data/20201006-maine Exo/(extracted) Lake-China catabot Sonde - KorEXO Measurement File Export - 100620 214150.csv'
CSV_FILE_PATH2 = '../Data/20201006-maine Exo/(extracted) Lake-China BlueRov Sonde - KorEXO Measurement File Export - 100620 214020.csv'


with open(CSV_FILE_PATH, 'r') as csv_file:
    reader = csv.reader(csv_file)
    tlog_list = np.array(list(reader))

# temp
z = tlog_list[1:4501,18]
z = z.astype('float32')

# GPS location
x = tlog_list[1:4501,25] 
x = x.astype('float64')
y = tlog_list[1:4501,24] 
y = y.astype('float64') 

f, ax = plt.subplots()
ax.set_title('Catabot China 10/06: Temperature (' + degree_sign + 'C)')
cs = ax.tricontourf(x,y,z, 10, norm=colors.SymLogNorm(linthresh=0.03, linscale=0.03))

f.colorbar(cs, orientation='horizontal', format='%.1f')
ax.set_xlabel('Longitude')
f.canvas.draw()

ax.set_ylabel('Latitude')
f.canvas.draw()

ax.plot(x,y,marker='o', color='k', markersize=0.1)
ax.tick_params(right= True,top= True,left= False, bottom= False)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_aspect('equal')
plt.grid(True)

bar = AnchoredSizeBar(ax.transData, 0.00046, '40 m', 4)
ax.add_artist(bar)
```
