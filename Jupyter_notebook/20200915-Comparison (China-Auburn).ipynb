```python
# Load essential libraries
import csv
import numpy as np
import matplotlib.pyplot as plt
import statistics 
import numpy as np
from scipy.signal import butter, lfilter, freqz
from IPython.display import Image

from datetime import datetime
```

```python
# File loading from relative path
file1 = '../Data/20200915-china.csv'
file2 = '../Data/20200915-auburn-entire.csv'
```

```python
# Figure initialization
fig = plt.figure()

################ DATA 1 ################
# Time and robot egomotion
time_1 = []
standardized_time_1 = []
compass_heading_1 = []

# sonde data
temp_1 = []
PH_1 = []
cond_1 = [] # ms
chlorophyll_1 = [] 
ODO_1 = [] # mg/L
sonar_1 = []
```

```python
################ DATA 2 ################
# Time and robot egomotion
time_2 = []
standardized_time_2 = []
compass_heading_2 = []

# sonde data
temp_2 = []
PH_2 = []
cond_2 = [] # ms
chlorophyll_2 = [] 
ODO_2 = [] # mg/L
sonar_2 = []
```

```python
with open(file1, 'r') as csvfile1:
    csvreader1= csv.reader(csvfile1, delimiter=',')
    header = next(csvreader1)
    for row in csvreader1:
        # robot data
        time_1.append(float(row[0]))
        compass_heading_1.append(float(row[4]))
        
        # sonde data
        temp_1.append(float(row[23]))
        PH_1.append(float(row[26]))
        cond_1.append(float(row[25]))
        chlorophyll_1.append(float(row[29]))
        ODO_1.append(float(row[31]))
        sonar_1.append(float(row[8]))

    minimum_time_1 = min(time_1)
    for time_stamp_1 in time_1:
        standardized_time_1.append(time_stamp_1 - minimum_time_1)
```

```python
with open(file2, 'r') as csvfile2:
    csvreader2= csv.reader(csvfile2, delimiter=',')
    header = next(csvreader2)
    for row in csvreader2:
        # robot data
        time_2.append(float(row[0]))
        compass_heading_2.append(float(row[4]))
        
        # sonde data
        temp_2.append(float(row[23]))
        PH_2.append(float(row[26]))
        cond_2.append(float(row[25]))
        chlorophyll_2.append(float(row[29]))
        ODO_2.append(float(row[31]))
        sonar_2.append(float(row[8]))

    minimum_time_2 = min(time_2)
    for time_stamp_2 in time_2:
        standardized_time_2.append(time_stamp_2 - minimum_time_2)
```

### Compass heading

```python
plt.plot(standardized_time_1, compass_heading_1, label='China')
plt.plot(standardized_time_2, compass_heading_2, label='Auburn')

plt.xlabel('Time [sec]', fontsize=16)
plt.ylabel('Heading [degree]', fontsize=16)
plt.legend()
plt.show()
```

### Temperature

```python
plt.plot(standardized_time_1, temp_1, label='China')
plt.plot(standardized_time_2, temp_2, label='Auburn')

plt.xlabel('Time [sec]', fontsize=16)
plt.ylabel('Temperature [degree]', fontsize=16)
plt.legend()
plt.show()
```

### PH

```python
plt.plot(standardized_time_1, PH_1, label='China')
plt.plot(standardized_time_2, PH_2, label='Auburn')

plt.xlabel('Time [sec]', fontsize=16)
plt.ylabel('PH', fontsize=16)
plt.legend()
plt.show()
```
